# Scipy has a wrong tail behaviour for |beta| == 1

Here is an example. In blue is the theoretical tail behaviour (implemented by levy-stable-jax), in orange scipy

```python
(alpha, beta) = (1.6, 1.0)
xs = np.linspace(-15, 10, 100)
sns.lineplot(x=xs,
             y=lsj.logpdf(xs, alpha,beta, param="N0"))
sns.lineplot(x=xs,
             y=sp_levy_stable.logpdf(xs,alpha=alpha, beta=beta))
```
